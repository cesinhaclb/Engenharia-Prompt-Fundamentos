Criar um programa que organize arquivos bagunçados em pastas
automaticamente, baseando-se apenas na extensão do arquivo (ex: . pdf, .jpg).
utilizer biblioteca os  ou shutil


```python
import os
import shutil

def organize_files_by_extension(source_directory):
    if not os.path.isdir(source_directory):
        print(f"Erro: O diretório '{source_directory}' não existe.")
        return

    print(f"Organizando arquivos em: {source_directory}")

    for filename in os.listdir(source_directory):
        file_path = os.path.join(source_directory, filename)

        # Ignora diretórios e links simbólicos
        if os.path.isfile(file_path):
            # Obtém a extensão do arquivo
            _, file_extension = os.path.splitext(filename)
            file_extension = file_extension.lower() # Converte para minúsculas

            # Se não houver extensão (ex: 'Makefile'), pode ser tratado como 'no_extension' ou ignorado
            if not file_extension:
                destination_folder = os.path.join(source_directory, "no_extension")
            else:
                # Remove o ponto da extensão para usar como nome da pasta
                destination_folder = os.path.join(source_directory, file_extension[1:])

            # Cria a pasta de destino se ela não existir
            os.makedirs(destination_folder, exist_ok=True)

            # Move o arquivo para a pasta de destino
            try:
                shutil.move(file_path, destination_folder)
                print(f"Movido '{filename}' para '{destination_folder}'")
            except Exception as e:
                print(f"Erro ao mover '{filename}': {e}")

    print("Organização de arquivos concluída!")

# Solicita ao usuário o diretório a ser organizado
source_dir = input("Digite o caminho do diretório que deseja organizar: ")
organize_files_by_extension(source_dir)
```

Desenvolver um script que consulte uma API publica (ViaCEP) e
retorne as informagées de endereco formatadas de maneira legivel.

```python
import requests

def consultar_cep(cep):
    """
    Consulta a API ViaCEP para obter informações de endereço.
    """
    # Remove caracteres não numéricos do CEP
    cep = ''.join(filter(str.isdigit, cep))

    if len(cep) != 8:
        return {"erro": True, "mensagem": "CEP inválido. Deve conter 8 dígitos."}

    url = f"https://viacep.com.br/ws/{cep}/json/"
    try:
        response = requests.get(url)
        response.raise_for_status()  # Levanta um erro para status de resposta HTTP ruins (4xx ou 5xx)
        data = response.json()

        if "erro" in data and data["erro"] is True:
            return {"erro": True, "mensagem": "CEP não encontrado."}
        return data
    except requests.exceptions.RequestException as e:
        return {"erro": True, "mensagem": f"Erro na conexão: {e}"}
    except ValueError:
        return {"erro": True, "mensagem": "Resposta inválida da API."}

def formatar_endereco(endereco):
    """
    Formata as informações de endereço para exibição legível.
    """
    if endereco.get("erro"):
        return f"Erro: {endereco['mensagem']}"

    logradouro = endereco.get("logradouro", "N/A")
    bairro = endereco.get("bairro", "N/A")
    localidade = endereco.get("localidade", "N/A")
    uf = endereco.get("uf", "N/A")
    cep = endereco.get("cep", "N/A")

    return (
        f"CEP: {cep}\n"
        f"Logradouro: {logradouro}\n"
        f"Bairro: {bairro}\n"
        f"Cidade/UF: {localidade}/{uf}"
    )

# --- Exemplo de Uso ---
if __name__ == "__main__":
    cep_digitado = input("Digite o CEP (apenas números): ")

    print("\nConsultando CEP...")
    dados_endereco = consultar_cep(cep_digitado)

    print("\n--- Informações do Endereço ---")
    print(formatar_endereco(dados_endereco))
```

Criar um programa autônomo que dispare alertas (mensagens de alerta no console
ou e-mail simulado) unicamente quando uma condição crítica for atingida.

```python
# Definimos um limite crítico para a temperatura
TEMPERATURA_CRITICA = 30

def verificar_condicao_critica(valor_monitorado):
    """
    Verifica se o valor monitorado atingiu uma condição crítica.
    Neste exemplo, é uma temperatura acima de TEMPERATURA_CRITICA.
    """
    return valor_monitorado > TEMPERATURA_CRITICA

def disparar_alerta(mensagem):
    """
    Dispara um alerta, imprimindo uma mensagem no console.
    Em um sistema real, poderia enviar um e-mail, SMS, etc.
    """
    print(f"\n{'='*30}")
    print(f"!!! ALERTA CRÍTICO: {mensagem} !!!")
    print(f"{'='*30}\n")

def main():
    """
    Função principal que simula o monitoramento de uma condição crítica (temperatura).
    Solicita a temperatura ao usuário e dispara um alerta se ela exceder o limite crítico.
    """
    print("Iniciando monitoramento de sistema autônomo...")
    print(f"Limite crítico de temperatura: {TEMPERATURA_CRITICA}°C")

    # Simulação de leitura de temperatura (em um sistema real, viria de um sensor)
    try:
        temperatura_str = input("Digite a temperatura atual do sistema em °C para verificar: ")
        temperatura_atual = float(temperatura_str)

        print(f"Verificando temperatura: {temperatura_atual}°C")

        if verificar_condicao_critica(temperatura_atual):
            disparar_alerta(f"Temperatura ({temperatura_atual}°C) excedeu o limite crítico de {TEMPERATURA_CRITICA}°C!")
        else:
            print("Condição normal. Nenhum alerta disparado.")

    except ValueError:
        print("Entrada inválida. Por favor, digite um número para a temperatura.")
    except Exception as e:
        print(f"Ocorreu um erro: {e}")

    print("Monitoramento concluído para esta verificação.")

if __name__ == "__main__":
    main()
```
